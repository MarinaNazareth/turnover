Construir tabelas de frequências absolutas e relativas para as variáveis qualitativas.

```python
#gerando um df com a frequencia relativa e absoluta de cada coluna que é qualitativa (object)
def gerar_tabelas_frequencia(df, coluna):
    if df[coluna].dtype == 'O':  
        # Tabela de frequências absolutas
        frequencia_absoluta = df[coluna].value_counts()

        # Tabela de frequências relativas
        frequencia_relativa = df[coluna].value_counts(normalize=True)

        # Criando um DataFrame com as duas frequencias
        tabela_completa = pd.DataFrame({'Frequência Absoluta': frequencia_absoluta, 'Frequência Relativa': frequencia_relativa})

        return tabela_completa
    else:
        return None
```

```python
#iteração sobre cada coluna pra geraro dataframe com o nome tabela_nomedacoluna e printar o resultado
for coluna in base.columns:
    nome_tabela = f'tabela_{coluna}'
    tabela = gerar_tabelas_frequencia(base, coluna)
    
    if tabela is not None:
        globals()[nome_tabela] = tabela
        print(f"{nome_tabela}:")
        print(globals()[nome_tabela])
        print("\n")
```

## Análise de associação

A taxa de turnover é maior em algum grupo especifico?

```python
#função para pegar todas as categorias qualitativas e ver o turnover por segmentação
def calcular_turnover_para_todas_categorias(df, coluna_funcionarios_sairam):
    colunas_categoricas = df.select_dtypes(include='O').columns.tolist()
    colunas_categoricas = [coluna for coluna in colunas_categoricas if coluna != coluna_funcionarios_sairam]

    # Criar um dicionário para armazenar as tabelas de turnover para cada categoria
    tabelas_turnover = {}

    for coluna in colunas_categoricas:
        tabela_turnover = pd.DataFrame(index=df[coluna].unique())
        
        tabela_turnover['Total Funcionários'] = df.groupby(coluna).size()
        tabela_turnover['Funcionários que Saíram'] = df[df[coluna_funcionarios_sairam] == 'Sim'].groupby(coluna).size()
        tabela_turnover['Taxa de Turnover (%)'] = (tabela_turnover['Funcionários que Saíram'] / tabela_turnover['Total Funcionários']) * 100

        nome_tabela = f'tabela_turnover_{coluna}'
        tabelas_turnover[nome_tabela] = tabela_turnover

    return tabelas_turnover


```

```python
#execução do def e exibição do resultado
tabelas_turnover = calcular_turnover_para_todas_categorias(base, 'Funcionário_deixou_a_empresa')

for nome_tabela, tabela_turnover in tabelas_turnover.items():
    print(f"\n{nome_tabela}:")
    print(tabela_turnover)

```

```python
#após executar a função que calcula, pegar a maior % de turnover por categoria

resultados_finais = {}

for nome_tabela, tabela_turnover in tabelas_turnover.items():
    categoria_maior_taxa = tabela_turnover.loc[tabela_turnover['Taxa de Turnover (%)'].idxmax()]
    resultados_finais[nome_tabela] = {
        'coluna': nome_tabela.replace('tabela_turnover_', ''),
        'categoria_maior_taxa': categoria_maior_taxa.name,
        'taxa_maior_turnover': categoria_maior_taxa['Taxa de Turnover (%)']
    }

# Exibir os resultados finais
for coluna, resultado in resultados_finais.items():
    print(f"\nColuna: {resultado['coluna']}")
    print(f"Categoria com Maior Taxa de Turnover: {resultado['categoria_maior_taxa']}")
    print(f"Taxa de Turnover da Categoria com Maior %: {resultado['taxa_maior_turnover']:.2f}%")
```

O turnover tem relação com alguma variável quantitativa?

```python
# Lista de variáveis quantitativas

variaveis_quantitativas = ['Idade', 'Distância_do_trabalho', 'Salário', 'Qte_Empresas_Trabalhadas',
       'Perc_de_aumento', 'Qte_ações_da_empresa', 'Tempo_de_carreira',
       'Horas_de_treinamento', 'Tempo_de_empresa', 'Anos_no_mesmo_cargo',
       'Anos_desde_a_ultima_promocao', 'Anos_com_o_mesmo_chefe']

#tamanho da faixa de cada coluna
tamanhos_bin = [10,5,2000,2,4,1,5,1,5,4,2,2]

    
tabelas_turnover_por_variavel = {}


for variavel_quantitativa in variaveis_quantitativas:
    tamanho_bin = tamanhos_bin[variaveis_quantitativas.index(variavel_quantitativa)]
    coluna_faixa = f'{variavel_quantitativa}_Faixa_{tamanho_bin}'
    base[coluna_faixa] = pd.cut(base[variavel_quantitativa], bins=range(0, int(base[variavel_quantitativa].max()) + tamanho_bin, tamanho_bin), right=False)
    tabela_turnover_por_faixa = pd.DataFrame()
    tabela_turnover_por_faixa['Total_Funcionários'] = base.groupby(coluna_faixa).size()
    tabela_turnover_por_faixa['Funcionários_que_Saíram'] = base[base['Funcionário_deixou_a_empresa'] == 'Sim'].groupby(coluna_faixa).size().fillna(0)
    tabela_turnover_por_faixa['Taxa_de_Turnover(%)'] = (tabela_turnover_por_faixa['Funcionários_que_Saíram'] / tabela_turnover_por_faixa['Total_Funcionários']) * 100
    tabela_turnover_por_faixa = tabela_turnover_por_faixa.reset_index()
    tabela_turnover_por_faixa = tabela_turnover_por_faixa.sort_values(by=coluna_faixa, ascending=True)
    chave_unica = f'{variavel_quantitativa}_{tamanho_bin}'
    tabelas_turnover_por_variavel[chave_unica] = tabela_turnover_por_faixa


for chave, tabela in tabelas_turnover_por_variavel.items():
    print(f"\nTabela para {chave}:")
    print(tabela)
```

```python
# Iterar sobre as tabelas de turnover por faixa para cada variável
for chave, tabela in tabelas_turnover_por_variavel.items():
    print('\n')
    tabela.drop(['Funcionários_que_Saíram', 'Total_Funcionários'], axis=1)

    for coluna in tabela.columns.difference(['Taxa_de_Turnover(%)', 'Funcionários_que_Saíram', 'Total_Funcionários']):
        # Obter a categoria com a maior taxa de turnover para cada coluna
        categoria_maior_taxa = tabela.loc[tabela['Taxa_de_Turnover(%)'].idxmax()]
        # Acessar o nome da categoria e a taxa de turnover
        nome_categoria_maior_taxa = categoria_maior_taxa[coluna]
        taxa_turnover_maior = categoria_maior_taxa['Taxa_de_Turnover(%)']

        print(f"Coluna: {coluna}")
        print(f"Categoria com Maior Taxa de Turnover: {nome_categoria_maior_taxa}")
        print(f"Taxa de Turnover da Categoria com Maior %: {taxa_turnover_maior:.2f}%")

```
